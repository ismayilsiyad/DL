```python
import pandas as pd
```

```python
re = pd.read_csv('https://raw.githubusercontent.com/CloudThat-HPE/ComprehensiveTrainingOnMLandDL/main/Datasets/Real%20estate.csv')
```

```python
re.head()
```

```python
re.drop(['No','X1 transaction date'], axis=1, inplace=True)
```

```python
re.head()
```

```python
re.isnull().sum()
```

```python
X=re.drop('Y house price of unit area', axis=1)
y=re['Y house price of unit area']
```

```python
from sklearn.preprocessing import MinMaxScaler
for col in X:
    mm = MinMaxScaler()
    X[col] = mm.fit_transform(X[[col]])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
n = X_train.shape[1]
```

```python
model.add(Dense(10, activation='relu', input_shape=(n,)))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train,y_train, batch_size=30, epochs=500)
```

```python
pred  = model.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, pred)
```

```python
r2_score(y_test, pred)
```
